# ExampleOne

```python
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
```

```python
housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names)
df
```

```python
X_selected = df[['MedInc', 'AveRooms']]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_selected)
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(X_scaled)
df['Cluster'] = clusters
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_selected['MedInc'], X_selected['AveRooms'], c=clusters, cmap='plasma')
plt.xlabel('Ø¯Ø±Ø¢Ù…Ø¯ Ù…ØªÙˆØ³Ø·')
plt.ylabel('Ù…ÛŒØ§Ù†Ú¯ÛŒÙ† ØªØ¹Ø¯Ø§Ø¯ Ø§ØªØ§Ù‚â€ŒÙ‡Ø§')
plt.title('Ø®ÙˆØ´Ù‡â€ŒØ¨Ù†Ø¯ÛŒ Ù…Ù†Ø§Ø·Ù‚ Ù…Ø³Ú©ÙˆÙ†ÛŒ Ú©Ø§Ù„ÛŒÙØ±Ù†ÛŒØ§')
plt.show()
```

# ExampleTwo

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
df = pd.read_csv('Mall_Customers.csv')
df
```

```python
km = KMeans(n_clusters=5)
km.fit(df[['Annual_Income_(k$)', 'Spending_Score_(1-100)']])
df['lable'] = km.labels_
```

```python
sns.scatterplot(x=df['Annual_Income_(k$)'],
                y=df['Spending_Score_(1-100)'],
                hue=df['lable'],
                palette='summer')
```
